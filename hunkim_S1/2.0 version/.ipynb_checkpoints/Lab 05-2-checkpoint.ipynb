```python
import tensorflow as tf
import numpy as np

xy = np.loadtxt('data-03-diabetes.csv', delimiter=',', dtype=np.float32)
x_data = xy[:, 0:-1]
y_data = xy[:, [-1]]
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(1, activation='sigmoid', dtype='float32')
])

sgd = tf.optimizers.SGD(learning_rate=0.01)
model.compile(optimizer=sgd, 
              loss=tf.keras.losses.BinaryCrossentropy(),
              metrics=[])

model.fit(x_data, y_data, verbose=2, epochs=10001)
```

```python
predicted = tf.cast(model.predict(x_data) > 0.5, dtype = tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, y_data), dtype=tf.float32))
print("\nHypothesis : ", model.predict(x_data), "\nCorrect (Y) : ", y_data, "\nAccuracy : ", accuracy.numpy())
```
