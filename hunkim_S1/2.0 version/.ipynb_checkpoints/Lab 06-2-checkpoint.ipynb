```python
import tensorflow as tf
import numpy as np
```

```python
xy = np.loadtxt('data-04-zoo.csv', delimiter=',', dtype=np.float32)
x_data = xy[:, 0:-1]
y_data = xy[:, [-1]]
```

```python
y_data_one_hot = tf.keras.utils.to_categorical(y_data)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(7, activation='softmax', dtype='float32')
])

sgd = tf.optimizers.SGD(learning_rate=0.01)
model.compile(optimizer=sgd, 
              loss=tf.keras.losses.CategoricalCrossentropy(),
              metrics=['acc'])

model.fit(x_data, y_data_one_hot, verbose=2, epochs=2000)
```

```python
pred = tf.argmax(model.predict(x_data), 1)
for p, y in zip(pred, y_data.flatten()):
    print("[{}] Prediction : {} True Y : {}".format(p == int(y), p, int(y)))
```
